Task#4: Split an Image into individual color channels

```python
image = cv2.imread('../images/HamzaCR.jpeg')
B, G, R = cv2.split(image)
cv2.imshow('Blue Channel', B)
cv2.imshow('Green Channel', G)
cv2.imshow('Red Channel', R)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

Task#6: Convert a Color image to Grayscale (Manually)

```python
import numpy as np

img = cv2.imread('../images/HamzaCR.jpeg')
B, G, R = cv2.split(img)
gray_img = 0.299 * R + 0.587 * G + 0.114 * B
gray_img = np.uint8(gray_img)
cv2.imshow('Grayscale Image', gray_img)
cv2.waitKey(0)
cv2.destroyAllWindows()
```
